```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Dataset.csv')
```

```python
df.head()
```

```python
df.rename(columns={'Hipertension':'Hypertension','Handcap':'Handicap','No-show':'No_show'},inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
df['ScheduledDayC'] = df['ScheduledDay'].copy()
df['AppointmentDayC']=df['AppointmentDay'].copy()
df
```

```python
# Step 2: Extract datetime features
# Day of week of appointment (0=Monday, 6=Sunday)
df['AppointmentWeekday'] = df['AppointmentDayC'].dt.weekday
df['AppointmentMonth'] = df['AppointmentDay'].dt.month

# Weekend indicator
df['IsWeekend'] = df['AppointmentWeekday'].isin([5, 6]).astype(int)

# Day of week of scheduling
df['ScheduledWeekday'] = df['ScheduledDayC'].dt.weekday

# Hour of the scheduled time
df['ScheduledHour'] = df['ScheduledDayC'].dt.hour

# Show result
df.head()
```

```python
#feature engineering
# List of binary condition columns
medical_conditions = ['Hypertension', 'Diabetes', 'Alcoholism']  # example

# Create new feature: total number of conditions
df['Total_Conditions'] = df[medical_conditions].sum(axis=1)
```

```python
df.head()
```

```python
df['ScheduledDay'] = df['ScheduledDay'].dt.date
df['AppointmentDay'] = df['AppointmentDay'].dt.date
df.head()
```

```python
df['WaitingDays'] = (df['AppointmentDay'] - df['ScheduledDay']).apply(lambda x: x.days)
df.head(5)
```

```python
negative_count = (df['WaitingDays'] < 0).sum()
print(negative_count)
```

```python
df.shape
```

```python
df = df[df['WaitingDays'] >= 0]
```

```python
df.shape
```

```python
df['No_show'].value_counts()
```

```python
(df['Age'] < 0).sum()
```

```python
#removal of negative ages
df = df[df['Age'] >= 0]
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
#encoding categorical values
from sklearn.preprocessing import OneHotEncoder
df['No_show']=df['No_show'].map({'Yes':1,'No':0})
df['Gender'] = df['Gender'].map({'F': 0, 'M': 1})
df=pd.get_dummies(df,columns=['Neighbourhood'],drop_first=True)
```

```python
# Binning Age into categories
bins = [0, 12, 18, 35, 60, 100]  # you can adjust based on your data
labels = ['Child', 'Teen', 'Young_Adult', 'Adult', 'Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
# One-hot encode the Age_Group column
df = pd.get_dummies(df, columns=['Age_Group'], drop_first=True)
```

```python
df.head()
```

```python
df.columns
```

```python
#scaling age and waiting days
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
num_cols=['Age','WaitingDays']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.head()
```

```python
df = df.drop(columns=['PatientId', 'AppointmentID','ScheduledDay','AppointmentDay','AppointmentDayC','ScheduledDayC'])
```

```python
X = df.drop('No_show', axis=1)
y = df['No_show']  
```

```python
print(y.value_counts(normalize=True))
#to handle imbalanced data use smote,weights(80:20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
```

```python
# Step	Apply When?	Use on Test Data?
# Baseline Model	Before trying fixes	Yes
# SMOTE	Before training	❌ No (only train)
# Class Weights	During training	✅ Yes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

# 🔹 Logistic Regression - Baseline
print("🔹 Logistic Regression - Baseline")
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
y_pred_lr_prob = lr.predict_proba(X_test)[:, 1]  # 🔸 Probabilities for class 1
print(classification_report(y_test, y_pred_lr))

# 🔹 Random Forest - Baseline
print("🔹 Random Forest - Baseline")
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
y_pred_rf_prob = rf.predict_proba(X_test)[:, 1]  # 🔸 Probabilities for class 1
print(classification_report(y_test, y_pred_rf))

# 🔹 XGBoost - Baseline
print("🔹 XGBoost - Baseline")
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')  # Avoids warning
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb_prob = xgb_model.predict_proba(X_test)[:, 1]  # 🔸 Probabilities for class 1
print(classification_report(y_test, y_pred_xgb))
```

```python
print("🔹 Logistic Regression - Class Weight Balanced")
lr_weighted = LogisticRegression(class_weight='balanced', max_iter=1000)
lr_weighted.fit(X_train, y_train)
y_pred_lr_w = lr_weighted.predict(X_test)
y_pred_lr_w_prob = lr_weighted.predict_proba(X_test)[:, 1]  # 🔸 probability for class 1
print(classification_report(y_test, y_pred_lr_w))

print("🔹 Random Forest - Class Weight Balanced")
rf_weighted = RandomForestClassifier(class_weight='balanced')
rf_weighted.fit(X_train, y_train)
y_pred_rf_w = rf_weighted.predict(X_test)
y_pred_rf_w_prob = rf_weighted.predict_proba(X_test)[:, 1]  # 🔸 probability for class 1
print(classification_report(y_test, y_pred_rf_w))
```

```python
# The stratify=y parameter in train_test_split() is used to ensure that the train and test splits 
# have the same proportion of classes as the original dataset.
# SMOTE cannot handle datetime.date objects — it only works with numerical features

#applying smopte on training only
from imblearn.over_sampling import SMOTE

# Resample with SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# 🔹 Logistic Regression with SMOTE
print("🔹 Logistic Regression - SMOTE")
lr_smote = LogisticRegression(max_iter=1000)
lr_smote.fit(X_train_smote, y_train_smote)
y_pred_lr_s = lr_smote.predict(X_test)
y_pred_lr_s_prob = lr_smote.predict_proba(X_test)[:, 1]  # 🔸 Add this line
print(classification_report(y_test, y_pred_lr_s))

# 🔹 Random Forest with SMOTE
print("🔹 Random Forest - SMOTE")
rf_smote = RandomForestClassifier()
rf_smote.fit(X_train_smote, y_train_smote)
y_pred_rf_s = rf_smote.predict(X_test)
y_pred_rf_s_prob = rf_smote.predict_proba(X_test)[:, 1]  # 🔸 Add this line
print(classification_report(y_test, y_pred_rf_s))
```

# Build the DL Model with Dropout + BatchNorm

```python
scaler = StandardScaler()
X_train_dl = scaler.fit_transform(X_train)
X_test_dl = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

# Deep Learning model
dl_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_dl.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation='relu'),
    Dropout(0.2),

    Dense(1, activation='sigmoid')  # Binary classification
])

dl_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Training
history = dl_model.fit(X_train_dl, y_train,
                       epochs=100,
                       batch_size=32,
                       validation_split=0.2,
                       callbacks=[early_stop],
                       verbose=1)
```

```python
#predict and classification report
# Predict class probabilities
y_pred_dl_prob = dl_model.predict(X_test_dl).flatten()
y_pred_dl = (y_pred_dl_prob >= 0.5).astype(int)

from sklearn.metrics import classification_report
print("🔹 Deep Learning Model with Dropout & BatchNorm")
print(classification_report(y_test, y_pred_dl))
```

```python
#plot roc curve
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_dl_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'DL ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Deep Learning Model')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
#plot dl training accuracy and loss
plt.figure(figsize=(12,5))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#roc comparision
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Calculate FPR, TPR for all models
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_lr_prob)
roc_auc_lr = auc(fpr_lr, tpr_lr)

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_rf_prob)
roc_auc_rf = auc(fpr_rf, tpr_rf)

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_pred_xgb_prob)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)

fpr_dl, tpr_dl, _ = roc_curve(y_test, y_pred_dl_prob)
roc_auc_dl = auc(fpr_dl, tpr_dl)

# Plot all ROC curves
plt.figure(figsize=(10, 7))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {roc_auc_xgb:.2f})')
plt.plot(fpr_dl, tpr_dl, label=f'Deep Learning (AUC = {roc_auc_dl:.2f})')

# Plot diagonal reference line
plt.plot([0, 1], [0, 1], 'k--', lw=2)

# Final touches
plt.title('ROC Curve Comparison: LR vs RF vs XGB vs DL')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

```
